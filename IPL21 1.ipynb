```python
plt.figure(figsize=(14,6))
a=np.arange(len(IPL21['Date']))

plt.plot(a+0.2,'Second Innings score',0.2,data=IPL21,label='Line 1', color='green')
plt.plot(a-0.2,'Second Innings Wicket',0.2,data=IPL21,label='Line 2', color='Black')
legend = plt.legend()
legend.get_frame().set_facecolor('black')  
plt.xlabel("Second Innings score")
plt.ylabel("Second Innings Wicket")
plt.title("Second Innings score VS Second Innings Wicket")
plt.xticks(a,IPL21['Date'],rotation='vertical')
plt.legend(['score','Wicket'])
plt.show()
```
